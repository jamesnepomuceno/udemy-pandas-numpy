## NUMPY SELECT

Lets you create columns based in multiple conditions

```python
oil = pd.read_csv('../retail/oil.csv', parse_dates=['date'], names=['date', 'price'], skiprows=1)
oil.head()
```

```python
oil.dtypes
```

```python
oil['buy'] = np.where(oil['price'] > 100, 'Too high', 'Buy')
oil.head()
```

```python
oil['buy'].value_counts()
```

```python
conditions = [
    (oil['price'] > 100),
    (oil['price'] > 50) & (oil['price'] <= 100),
    (oil['price'] <= 50)
]
choices = ['Too high', 'Buy', 'Strong buy']
oil['buy'] = np.select(conditions, choices, default='Missing')
oil.head()
```

### PRO-TIP: Column creation with .assign()
> Tempestividade


### Look at this example:

<code> ....assign( <br>
    tax = (sample_df['sales'] * 0.05).round(2).map(lambda x: f'${x})) <br>
    .query('year > 2017') <br>
</code>)

Cada execução dentro do assign retornará um DF, mas o dataframe completo <br>
só retornará no final da execução.

> DEMO

```python
retail = pd.read_csv('../retail/retail_2016_2017.csv')
sample_df = retail.sample(5, random_state=85)
sample_df
```

```python
sample_df.assign(
    oppromotion_flag = sample_df['onpromotion'] > 0,
    family_abbrev = sample_df['family'].str[:3],
    on_promotion_ratio = sample_df['sales'] / sample_df['onpromotion'],
    sales_onprom_target = lambda x: x['on_promotion_ratio'] > 100
).query('sales_onprom_target == True')
```

1. Observe que consigo incluir nas operações colunas que ainda não existiam.
2. Isso é possível porque a cada ASSIGN existe a criação de um DF intermediário
3. Com isso o próximo já consegue "visualizar" esse campo novo.

```python
## Também podemos atribuir isso a um novo DF (ou sobrescrever o anterior, claro)

df_1 = sample_df.assign(
    oppromotion_flag = sample_df['onpromotion'] > 0,
    family_abbrev = sample_df['family'].str[:3],
    on_promotion_ratio = sample_df['sales'] / sample_df['onpromotion'],
    sales_onprom_target = lambda x: x['on_promotion_ratio'] > 100
).query('sales_onprom_target == True')

df_1.head()
```
